```python
#removing the missing values. 
df.dropna(inplace=True)
```

```python
#Convert categorical to numerical
df_oceanProx=pd.get_dummies(df,columns=['ocean_proximity'],prefix='dmy',drop_first=True)

df_oceanProx
```

```python
#Preparing to train the data
from sklearn.model_selection import train_test_split
#df_oceanProx.info()
```

```python
#Preparing to train the data
from sklearn.model_selection import train_test_split

X_feat= df_oceanProx.drop(columns=['median_house_value'],axis=1)
y_tgt= df_oceanProx['median_house_value']
```

```python
#Allowing the data to split into test and traiining data
x_train,x_test,y_train, y_test= train_test_split(X_feat,y_tgt,test_size=0.3, random_state=101)
```

```python
#Adjust the size of the following graph
plt.figure(figsize=(20,6))
#Use a heat map to seet= correlations
sns.heatmap(data=df_oceanProx.corr(), annot=True, cmap='coolwarm')
```

```python
#sns.pairplot(data=df_oceanProx)
```

```python
#Standardize the data!!!!!
from sklearn.preprocessing import StandardScaler
#Create object
scale_model= StandardScaler()
#Fit and Transform_Update the training and test models
x_train=scale_model.fit_transform(x_train)
x_test= scale_model.transform(x_test)
```

```python
#Call the Linear Regression Model from the Sklibrary
from sklearn.linear_model import LinearRegression
#Instantiate the model
lrModel= LinearRegression()
```

```python
#Place the training Model
lrModel.fit(x_train,y_train)
```

```python
#Place Predict Model
Ypred=lrModel.predict(x_test).round()


```

```python
#Testing the model
from sklearn.metrics import mean_squared_error as mse

print(mse(y_test,Ypred,squared=False)) #Square rooted the value. What to compare this value with?
```

```python
#Import KNN regression- Predict a continous value
from sklearn.neighbors import KNeighborsRegressor
#Set as an object
knr_model = KNeighborsRegressor(n_neighbors=10)
```

```python
#Fit the data to the KNeighbours
knr_model.fit(x_train,y_train)
```

```python
knrYpredict= knr_model.predict(x_test)
```

```python
print('Validation Data:',mse(y_test,knrYpredict, squared= False))
print('')
print('Training Data: ',mse(y_train,knr_model.predict(x_train), squared= False))

#Notes: IF TRAINIG DATA IS GREATER THAN THE VALIDATION DATA (BY A BIG MARGIN), WE HAVE AN OVERFITTING ISSUE
```

```python
#Random Forest 
from sklearn.ensemble import RandomForestRegressor
rfr= RandomForestRegressor(max_depth=8)
```

```python
rfr.fit(x_train,y_train)
```

```python
rfrYpredict=rfr.predict(x_test)
```

```python
print('Validation Data:', mse(y_test,rfrYpredict, squared=False))
print('')
print('Training Data:', mse(y_train, rfr.predict(x_train), squared=False))

#At Max Depth=2, it did not do very well
#At max depth= 10, trainig data was far less than the validation data, it did okay
```
